##  Problem 1: Mobile Recharge Validation System

### Problem Statement
Create a function to validate a mobile recharge request.

### Rules
- Recharge amount must be ≥ ₹50  
- Valid plans: 199, 299, 399, 599  
- User should be allowed to retry if invalid  

### Real-Time Use
- Telecom recharge apps  
- Payment gateways 

```python
def valid_recharge(amount):
    valid_plans=[199,299,399,599]
    if amount<=50:
        return "invalid option"
    if amount not in valid_plans:
        return "recharge rejected"
    return "recharge successful"
    
    if amount<=50:
        return valid_recharge
    
while True:
    amount = int(input("Enter recharge amount/plans : "))
    
    result = valid_recharge(amount)
    print(result)
    
    if result == "recharge successful":
        break
    else:
        print("Please try again...\n")              
        
```

## Problem 2: Inventory Reorder Alert System

### Problem Statement
Given a dictionary of products and stock quantities:

- Print **"Reorder Alert"** if stock < 15  
- Print **"Stock OK"** otherwise  

### Real-Time Use
- Warehouse systems  
- Retail management

```python
def inventory_reorder_alert(products):
    result = {}

    for product in products:
        if products[product] < 15:
            result[product] = "Reorder Alert"
        else:
            result[product] = "Stock OK"

    return result


products = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 8,
    "Wheat": 20
}

alerts = inventory_reorder_alert(products)

for item, status in alerts.items():
    print(f"{item}: {status}")
```

##  Problem 3: Student Result Processing System

### Problem Statement
Create a function that:

- Accepts a list of student marks  
- Calculates average marks  
- Prints **Pass** if average ≥ 50 else **Fail**  

### Real-Time Use
- Online exam systems  
- Evaluation portals 

```python
def student_result(marks):
    total=0
    for m in marks:
        total +=m
        
    average=total/len(marks)
    
    if average>=50:
        return "Pass"
    return "Fail"
no_of_marks=int(input("enter number of marks"))

for i in range(no_of_marks):
    marks=[int(input("enter marks :"))]
    
result=student_result(marks)
print(result)
```

##  Problem 4: Cab Fare Estimator with Retry Option

### Problem Statement
Create a cab fare calculator:

- Base fare = ₹50  
- ₹12 per km  
- Peak hour → 25% extra  
- Allow retry using `while` loop  

### Real-Time Use
- Ride-hailing apps  
- Transport systems  

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (distance * 12)
    
    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra
    
    return fare


# Retry using while loop
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")
    
    total_fare = calculate_fare(distance, peak)
    
    print("Total Fare = ₹", total_fare)
    
    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

##  Problem 5: Employee Attendance Eligibility Checker

### Problem Statement
Create a function that:

- Accepts attendance list (`"P"`, `"A"`)  
- Calculates attendance percentage  
- Returns **Eligible** if ≥ 75% else **Not Eligible**  

### Real-Time Use
- HR payroll systems  
- Internship tracking 

```python
def check_eligibility(attendance_list):
    present = 0
    
    for day in attendance_list:
        if day.upper() == "P":
            present += 1
    
    percentage = (present / len(attendance_list)) * 100
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Taking user input
attendance_list = []

n = int(input("Enter total number of working days: "))

for i in range(n):
    status = input("Enter attendance (P/A): ")
    attendance_list.append(status)

# Function call
result = check_eligibility(attendance_list)

print("Attendance Status:", result)  
```

##  Problem 6: Password Strength Checker

### Problem Statement
Create a function to check password strength.

### Rules
- Minimum length = 8  
- Must contain at least one digit  
- Must contain at least one special character (`@`, `#`, `$`)  

### Real-Time Use
- Login systems  
- Security validation 

```python
def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False
    
    if len(password) < 8:
        return False
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    return has_digit and has_special


# Retry until strong password
while True:
    pwd = input("Enter password: ")
    
    if check_password(pwd):
        print("Strong Password ✅")
        break
    else:
        print("Weak Password ❌")
        print("Password must:")
        print("- Be at least 8 characters long")
        print("- Contain at least one digit")
        print("- Contain at least one special character (@, #, $)")
        print("Please re-enter a strong password.\n")
```
