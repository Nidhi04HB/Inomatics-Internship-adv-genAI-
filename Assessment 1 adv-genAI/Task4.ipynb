## Task 4: Message Length Analyzer

### Problem Statement
Analyze message sizes.

### Given Data
```python
messages = ["Hi", "Welcome to the platform", "OK"]


````python
# Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)
    
    if length > 10:
        print("Flag: Message longer than 10 characters")
    
    print()  # for spacing
````
