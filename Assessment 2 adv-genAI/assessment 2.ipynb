## Problem 1: Unique Words in a Sentence

### Problem Statement
Given a sentence, identify all unique words and display:
- Count of unique words
- Unique words themselves

### Input
```python
sentence = "python is easy and python is powerful"


````python


sentence = input("Enter sentence: ")

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
````

## Problem 2: Highest Salary from Employee Data

### Problem Statement
Determine employee with highest salary.



````python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = ""
highest_salary = 0

for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

print("Highest Salary:", highest_employee, "-", highest_salary)
````

## Problem 3: Find Maximum and Minimum Values

### Problem Statement
Find maximum and minimum numbers from list.




````python
numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
````

## Problem 4: Count Products Above Price Threshold

### Problem Statement
Count products priced above 1000.



````python
prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
````

## Problem 5: Calculate Attendance Percentage

### Problem Statement
Calculate attendance percentage.

````python
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
````

## Problem 7: Count Character Frequency

### Problem Statement
Count frequency of each character.

````python
text = "python"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
````

## Problem 10: Calculate Average Marks

### Problem Statement
Calculate average marks.


````python
marks = [70, 80, 65, 75]

total = 0

for mark in marks:
    total += mark

average = total / len(marks)

print("Average Marks:", average)
````
