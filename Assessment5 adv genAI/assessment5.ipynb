##  Problem 1: Social Media – Post Engagement Analyzer

Create a function that analyzes engagement on a social media post.  

- How will you calculate the total number of likes?  
- How will you determine whether the post is **Viral** or has **Normal Engagement**?  
- If total likes are greater than or equal to 1000, classify it as *Viral Post*.  
- Otherwise, classify it as *Normal Engagement*.  


```python
def analyze_post_engagement(likes_list):
    total_likes = 0
    
    # Calculate total likes using loop
    for like in likes_list:
        total_likes += like
    
    print("Total Likes:", total_likes)
    
    # Check Viral or Normal
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example Data
likes = [200, 150, 300, 400]   # Total = 1050

analyze_post_engagement(likes)
```

## Problem 3: Agriculture – Rainfall Adequacy Checker

Create a function that analyzes seasonal rainfall data.  

- How will you calculate the average rainfall from given values?  
- If the average rainfall is greater than or equal to the required level, classify it as *Adequate Rainfall*.  
- Otherwise, classify it as *Inadequate Rainfall*.  



```python
def rainfall_checker(rainfall_data, required_level):
    total = 0
    
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    
    print("Average Rainfall:", average)
    
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

# Example
rainfall_checker([70, 75, 80, 63], 70)
```

##  Problem 4: Social Media – Duplicate Account Detection

Create a function that detects duplicate usernames in a system.  

- How will you check if duplicate usernames exist?  
- If duplicates are found, display *Duplicate Accounts Found: Yes*.  
- If no duplicates exist, display *Duplicate Accounts Found: No*.  


```python
def detect_duplicate_accounts(usernames):
    unique_users = set(usernames)
    
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

# Example
detect_duplicate_accounts(["user1", "user2", "user1"])
```

## Problem 6: Agriculture – Premium Crop Price Filter

Create a function that filters crop prices above a premium threshold.  

- If price is greater than ₹2000, classify it as *Premium Crop*.  
- Otherwise, classify it as *Regular Crop*.  


```python
def premium_crop_filter(prices):
    premium_crops = []
    
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)

# Example
premium_crop_filter([1500, 2500, 3200, 1800])
```

##  Problem 9: E-Learning – Student Attendance Eligibility System

Create a function that calculates attendance percentage and checks exam eligibility.  

- If attendance percentage is greater than or equal to 75%, display *Eligible*.  
- Otherwise, display *Not Eligible*.  


```python
def attendance_eligibility(attendance):
    present_count = 0
    
    for day in attendance:
        if day == "P":
            present_count += 1
    
    percentage = (present_count / len(attendance)) * 100
    
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

# Example
attendance_eligibility(["P", "P", "A", "P", "P"])
```


##  Problem 10: Smart Electricity Bill Analyzer

Create a function that calculates the electricity bill and classifies usage level based on units consumed.  

- First 100 units → ₹3 per unit  
- Next 100 units (101–200) → ₹5 per unit  
- Above 200 units → ₹7 per unit  

After calculating the total bill:

- If bill is less than ₹500 → classify as *Low Usage*.  
- If bill is between ₹500 and ₹1500 → classify as *Moderate Usage*.  
- If bill is above ₹1500 → classify as *High Usage*.  


```python
def electricity_bill(units):
    bill = 0
    
    # Billing calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    return bill, status


# Example
total_bill, usage_status = electricity_bill(250)

print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
